## Cross Validation, Regularization and Hyperparameter Optimization
###
## Cross Validation
* Cross-Validation is a technique for validating the model efficiency by training it on the subset if input data and testing on previously unseen subset.
* It is used to estimate the skill of model on new data.
* We can also say that it is a technique to check how a statistical model generalizes to an independent dataset.
######
### Regularization
- Regularization is a technique to avoid overfitting of the model.
- It is used to penalize the model for having high complexity.
######
### Hyperparameter Optimization
* It is a technique to find the best set of hyperparametres for the model.
* It is used to find the best set of hyperparametres for the model.
